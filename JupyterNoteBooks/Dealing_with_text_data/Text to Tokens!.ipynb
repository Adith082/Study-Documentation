```python
# going to implement a simple text tokenizer
# the vocabulary(means map of token(word, punctuation as String) and tokenId) of that tokenizer is already
# created from that short story!
# now we will structure our simple text tokenizer along with two functions called encode and decode
# encode will provide a list of tokenId from any given input text(of course token has tobe within instantiated vocabulary)
# decode will do the vice versa of the encode
# let's head to the implementation section!
```

```python
class simpleTokenizerVersion1:
    def __init__(self, vocabulary):
        self.str_to_int_map = vocabulary # here vocabulary, str_to_int_map and int_to_str_map all are dictionary
        self.int_to_str_map = {i:s for s,i in vocabulary.items()}
    def encode(self, text): # provides tokenids for each tokens of the given text
        preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', text)
        preprocessed = [item for item in preprocessed if item.strip()]
        token_ids = [self.str_to_int_map[token] for token in preprocessed]
        return token_ids
    def decode(self, token_ids):
        text = " ".join([self.int_to_str_map[token_id] for token_id in token_ids])
        #now we wanna remove the spaces before each punctuation
        text = re.sub(r'\s+([,.:;?_!"()\'])', r'\1', text) #here sub means substitution
        return text
```

```python
# lets test the above simple tokenizer version1 class
tokenizer = simpleTokenizerVersion1(vocab) #vocab is from the above short story
token_ids = tokenizer.encode("It's the last he painted, you know")
print(token_ids)
```

```python
print(tokenizer.decode(token_ids)) # decoded from tokenids to text
```

```python
# now what about when trying to encode, the tokenizer might not be able to
# find a particular token in its own vocabulary. In that case, the above tokenzer class object would throw
# error. but in order to mitigate the situation, we can introduce special context tokens.
# 2 special tokens will be '<|unk|>' and '<|endoftext|>' 
# here, unk token will be placed for any unknown words found and endoftext will be set for differentiating two different document texts

all_tokens = sorted(list(set(preprocessed))) #removing the duplicate tokens and converting it to a list and sort the unique tokens
# now we extend the above all_tokens with special context tokens
all_tokens.extend(["<|unk|>", '<|endoftext|>'])
# now its time to create the vocabulary
vocab = {token:token_id for token_id, token in enumerate(all_tokens)} # it is a dictionary
print(len(vocab))
```

```python
#printing the last 5 items of vocabulary in 2 different ways
vocab_items_list = list(vocab.items())
print(vocab_items_list[-5:])
print("")
for item in vocab_items_list[-5:]:
    print(item)
```

```python
class SimpleTokenizerVersion2: # this version contains 2 additional context tokens
    def __init__(self, vocabulary):
        self.str_to_int_map = vocabulary # here vocabulary, str_to_int_map and int_to_str_map all are dictionary
        self.int_to_str_map = {i:s for s,i in vocabulary.items()}
    def encode(self, text): # provides tokenids for each tokens of the given text
        preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', text)
        preprocessed = [item for item in preprocessed if item.strip()]
        preprocessed = [item if item in self.str_to_int_map else "<|unk|>" for item in preprocessed]
        token_ids = [self.str_to_int_map[token] for token in preprocessed]
        return token_ids
    def decode(self, token_ids):
        text = " ".join([self.int_to_str_map[token_id] for token_id in token_ids])
        #now we wanna remove the spaces before each punctuation
        text = re.sub(r'\s+([,.:;?_!"()\'])', r'\1', text) #here sub means substitution
        return text
```

```python
# time to test version 2 tokenizer
text1 = "hi there beautiful"
text2 = "please dont hurt me"
text = " <|endoftext|> ".join((text1, text2)) # two texts are separated with <|endoftext|>

tokenizer = SimpleTokenizerVersion2(vocab) # as you can see hi and beautiful are unknown to the vocavulary thats why it got replaced with <|unk|>
print(tokenizer.encode(text))
print(tokenizer.decode(tokenizer.encode(text))) # you can see the tokens here and how unknown tokens are defined and how endoftext tokens are here
```
